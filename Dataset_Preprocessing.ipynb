```python
from datasets import load_dataset

ds = load_dataset("cais/mmlu", "anatomy", split='test')
import pandas as pd

# Assuming df is the DataFrame loaded from your dataset
df = ds
def add_combined_str(example):
    # Ensure 'choices' is iterable (list or already valid list-like structure)
    if isinstance(example['choices'], list):
        # Combine the first four choices into a single string (or as many as available)
        combined_str = ' '.join(example['choices'][:4])  # Take up to the first 4 choices
    else:
        combined_str = ''

    # Add the combined string as a new key
    example['str_choices'] = combined_str
    return example

def add_correct_answer(example):
    # Ensure 'choices' is a list and 'answer' is a valid index
    if isinstance(example['choices'], list) and 0 <= example['answer'] < len(example['choices']):
        # Get the correct choice based on the 'answer' index
        correct_choice = example['choices'][example['answer']]
    else:
        correct_choice = None  # Handle cases where data is invalid

    # Add the correct choice as a new key
    example['correct_answer'] = correct_choice
    return example


# Apply the function to the dataset using map
df = df.map(add_combined_str)
df = df.map(add_correct_answer)

# The dataset now has an additional column str_choices
print(df)
```
